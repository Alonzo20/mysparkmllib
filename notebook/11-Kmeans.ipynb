```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataMat = []
```

```python
base_path = "/Users/zhangjianfeng/workspaces/workspace_github_bg/mysparkmllib/data/Kmeans/"
fr = open(base_path+"testSet.txt")
```

```python
for line in fr.readlines():
    curLine = line.strip().split('\t')
    fltLine = list(map(float,curLine))
    dataMat.append(fltLine)
```

```python
# 训练模型
km = KMeans(n_clusters=4) # 初始化
km.fit(dataMat) # 拟合
km_pred = km.predict(dataMat) # 预测
centers = km.cluster_centers_ # 质心
```

```python

# 可视化结果
plt.scatter(np.array(dataMat)[:, 1], np.array(dataMat)[:, 0], c=km_pred)
plt.scatter(centers[:, 1], centers[:, 0], c="r")
plt.show()
```
